```python
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
img_bgr = cv2.imread("29.jpg")
img_mask = cv2.imread("test_result.png")
```

```python
img_mask = img_mask[:,:,1]
palette_dict = {0:(0, 0, 0),
                255:(0,0,255)}
# 将整数ID，映射为对应类别的颜色
viz_mask_bgr = np.zeros((img_mask.shape[0], img_mask.shape[1], 3))
for idx in palette_dict.keys():
    print(np.where(img_mask==idx))
    viz_mask_bgr[np.where(img_mask==idx)] = palette_dict[idx]
viz_mask_bgr = viz_mask_bgr.astype('uint8')
```

```python
opacity = 0.5 # 透明度，越大越接近原图
img_bgr = cv2.addWeighted(img_bgr, opacity, viz_mask_bgr, 1-opacity, 0)
```

```python
cv2.imwrite("./result.jpg", img_bgr)
```
